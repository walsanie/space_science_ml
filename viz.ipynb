```python
import matplotlib.pyplot as plt
```

```python
plt.bar(highest_ten.index, highest_ten)
plt.xticks(rotation=90)
plt.ylabel('No. of Missions')
plt.show()
```

Number of launches in 2020 per company.

```python
df[df['Datum'].str.contains('2020')]['Company Name'].value_counts()
```

```python
hbar = df[df['Datum'].str.contains('2020')]['Company Name'].value_counts()
plt.barh(hbar.index, hbar)
plt.xlabel('No. Missions in 2020')
plt.xticks(range(1,21))
plt.show()
```

Percentage of successful and failed missions.

```python
df['Status Mission'].value_counts()
```

```python
per = df['Status Mission'].value_counts()
plt.pie(per, labels=per.index, autopct='%1.1f%%')
```

Percentages of Active and Retired rockets.

```python
act_ret = df[df['Company Name'].str.contains('SpaceX')]['Status Rocket'].value_counts()
plt.pie(act_ret, labels=act_ret.index, autopct='%1.1f%%')
```

Plotting the number of missions over the years.
<br>
Extract the year from the Datum column and count the rows per year.

```python
years = df['Datum'].str.replace(r'.*,\s+([0-9][0-9][0-9][0-9]).*', 
                                r'\1', 
                                regex=True).value_counts().sort_index().rename(index=int)
years
```

```python
print(years.index.dtype)
plt.plot(years)
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.xticks(range(1957, 2021, 10))
plt.xticks(rotation=90)
plt.show()
```

```python
company_mission_status = df[(df['Status Mission'] == 'Failure') | 
                        (df['Status Mission'] == 'Prelaunch Failure') | 
                        (df['Status Mission'] == 'Partial Failure')][['Company Name', 'Status Mission']].value_counts()[:16]
company_mission_status
```

```python
plt.scatter(company_mission_status.index.get_level_values(1), 
            company_mission_status.index.get_level_values(0), 
            s=company_mission_status*10, alpha=0.5)
plt.xticks(rotation=90)
plt.ylabel('No. of Missions')
plt.figure(figsize=(10,20))
plt.show()
```
